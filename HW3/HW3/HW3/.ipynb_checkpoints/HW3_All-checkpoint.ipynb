## Task I-b: Improve Schema

- There are several problems with the raw characters and actors information. Some obvious examples are:

    - The columns are not typed.

- Create a schema ```HW3_GOT_Fixed``` that has an improved schema and data model. Show your create and alter table, and data loading statements below. Also, run a query against your tables to show the data.

__Do not pull an Aly.__

````python
%sql drop schema if exists HW3_GOT_Fixed
````

````python
%sql create schema HW3_GOT_Fixed
````

````python
%%sql

    drop table if exists HW3_GOT_Fixed.characters;
    
    create table HW3_GOT_Fixed.characters as
        select characterName, characterLink, id, royal, kingsguard,
                nickname, characterImageThumb, characterImageFull
            from HW3_GOT_RAW.characters;
            
````

````python
%%sql

use HW3_GOT_Fixed;

alter table characters modify characterName varchar(128) null;

alter table characters modify characterLink varchar(64) null;

alter table characters modify id varchar(32) null;

alter table characters modify royal boolean null;

alter table characters modify kingsguard boolean null;

alter table characters modify nickname varchar(128) null;

alter table characters modify characterImageThumb varchar(1024) null;

alter table characters modify characterImageFull varchar(1024) null;
````

````python
%%sql

alter table characters
	add constraint characters_pk
		primary key (id);
````

````python
%%sql

    update characters
        set characterLink =
            substr(characterLink,
                length('/characters/'),
                length(characterLink)-length('/characters')-1);
            
````

````python
%%sql

    drop table if exists HW3_GOT_Fixed.actors;
    
    create table HW3_GOT_Fixed.actors as
        select actorName, actorLink
            from HW3_GOT_RAW.characters where actorName is not NULL;
````

- Current create statement.


````python
%%sql
    update actors
        set actorLink = substr(actorLink, length("/name/")+1, length(actorLink)-length("/name")-2);
````

- I am going to use an associative entity to handle the case in which the character-actor relationships might be many to many.


- There query below creates the basic information for exploration.


- __Do not pull an Aly.__

````python
%%sql


select * from
    characters join actors
where
    exists (select * from HW3_GOT_Raw.characters a
        where a.characterName=characters.characterName and actors.actorName=a.actorName);
````

- Build and cleanup the associative entity.


````python
%%sql

drop table if exists HW3_GOT_Fixed.characters_actors;

create table HW3_GOT_Fixed.characters_actors as
select characters.id as character_id, actors.id as actor_id from
    characters join actors
where
    exists (select * from HW3_GOT_Raw.characters a
        where a.characterName=characters.characterName and actors.actorName=a.actorName);
````

````python
%%sql

alter table characters_actors
	add constraint characters_actors_pk
		unique (character_id, actor_id);
````

````python
%%sql

alter table characters_actors
	add constraint characters_actors_fk2
		foreign key (actor_id) references actors (id);

alter table characters_actors
	add constraint characters_actors_characters_id_fk
		foreign key (character_id) references characters (id);
````

# Task II: Relationships

## Task II-a: Getting Relationship Data

- The MongoDB collection for ```characters``` has fields representing one-to-many relationships between characters.


- The fields are in the list below.

````python
relationship_names = [
    'abducted',
 'abductedBy',
 #'actors',
 'allies',
 'guardedBy',
 'guardianOf',
 'killed',
 'killedBy',
 'marriedEngaged',
 'parentOf',
 'parents',
 'servedBy',
 'serves',
 'sibling',
 'siblings'
]
````

- The Task II-a objective is to produce a table ```HW3_GOT_Raw.character_relationships``` of the form:

```character_relationships(sourceCharacterName, relationship, targetCharacterName)```


- Producing this information requires some pretty tricky MongoDB aggregate pipeline development. The critical hint is to realize that:
    - You can write a function that implements a generic pipeline to produce the information given a specific relationship name.
    - Write a python function that saves the information produced by the function in the SQL table.
    - Write a python loop that calls the function to produce the information for each of the relationships in the list above and saves/appends the information to the relationship table.

- __Without pulling an Aly, put your function here.__

````python
def get_relationships(property_name):

    client = MongoClient('mongodb://localhost:27017/')

    relationship_selector = '$' + property_name

    filter_template = [
            {
                '$unwind': {
                    'path': relationship_selector
                }
            }, {
                '$project': {
                    'source_character_id': {
                        '$toString': '$_id'
                    }, 
                    '_id': 0, 
                    'sourceCharacterName': '$characterName', 
                    'relationship': property_name, 
                    'targetCharacterName': relationship_selector
                }
            }
        ]
    
    result = client.HW3.GOT_Characters.aggregate(filter_template)
    
    return list(result)

````

````python
all_relationships = []

for r in relationship_names:
    tmp = get_relationships(r)
    all_relationships.extend(tmp)
````

````python
all_relationships
````

````python
relationships_df = pd.DataFrame(all_relationships)
````

````python
relationships_df.to_sql('character_relationships', con=engine,
                            schema='HW3_GOT_Raw', index=False, if_exists="replace")
````

````python
%%sql
drop table if exists HW3_GOT_Fixed.character_relationships;

create table HW3_GOT_Fixed.character_relationships as
    select
        cr.source_character_id, cr.sourceCharacterName,
        relationship,
        c.id as target_character_id,
        c.characterName as targetCharacterName
    from HW3_GOT_Raw.character_relationships as cr
        join
        characters as c
        on
        cr.targetCharacterName=c.characterName
````

````python
%%sql

alter table character_relationships modify source_character_id varchar(32) null;

alter table character_relationships modify sourceCharacterName varchar(128) null;

alter table character_relationships modify relationship varchar(32) null;
````

````python
%%sql

alter table character_relationships
	add constraint character_relationships_pk
		primary key (source_character_id, target_character_id, relationship);
````

````python
%%sql

alter table character_relationships
	add constraint character_relationships___fk2
		foreign key (target_character_id) references characters (id);

alter table character_relationships
	add constraint character_relationships_characters_id_fk
		foreign key (source_character_id) references characters (id);
````
